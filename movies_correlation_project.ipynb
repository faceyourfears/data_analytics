```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy  as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plots we will create

# Read in the data

df = pd.read_csv('/Users/lovesol/Downloads/movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
#Let's see if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
#Dropping all rows with missing data

df = df.dropna()
```

```python
#Change data type of the columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
#To show starting and ending data

df.head()
```

```python
#Order by gross revenue

df.head().sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Show the first 10 rows 
#Config type command

pd.set_option('display.max_rows', 10)

#Skips all the files in between
```

```python
#Show all the data
#Config type command

pd.set_option('display.max_rows', 50)

#If you look at the table there are dotted skip of data right in tghe middle
#By setting the parameter off, it'll show all the data, by not skipping any of it
```

```python
#Drop any duplicates
df.drop_duplicates()
```

Comparing budget and gross and their correlation

```python
#Scatter plot with budget vs gross 
plt.scatter(x = df['budget'], y = df['gross'])

#Giving the graph title
plt.title('Budget vs Gross Earnings')

#Lbaeling x and y axis
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

#Show the graph
plt.show()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

Let's start looking at the correlation

```python
# Methods (Pearson, Kendall, Spearman)
# Default is Pearson

df.corr(method = 'pearson')
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Company

df.head()
```

```python
#So you can numerize all the object type data


df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# After changing the data type, it looks like this with the correlation


correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Just by matrix

df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
